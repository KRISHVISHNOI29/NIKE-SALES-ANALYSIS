## CLEANING AND PREPROCESSING

### CLEANING AND PREPROCESSING THE SIZE COLUMN

```python
# DISPLAYS UNIQUE VALUES AND THEIR COUNT TO KNOW ABOUT DIVERSITY AND ENTRIES IN THE SIZE COLUMN

data['Size'].value_counts()
```

```python
data[data['Size'].isin(['M',"L","XL"])]
```

```python
# REPLACING THE SIZES FIRSTLY IN STRING TO MATCH THE PRESET DATATYPE AND THEN WE'LL CONVERT IT INTO THE INT64

data['Size'].replace("M","8",inplace=True)
data['Size'].replace("L","10",inplace=True)
data['Size'].replace("XL","12",inplace=True)
```

```python
data["Size"].value_counts()
```

```python
# ONE OF THE TWO ENTRIES IN THE SIZE COLUMN IS A STRING AND THE OTHER IS AN INT64, SO WE'LL CONVERT THE DATATYPE OF THE SIZE COLUMN FROM OBJECT TO INT64

data[data['Size']=="8"].count() 
```

```python
# CHANGING THE DATATYPE OF THE SIZE COLUMN FROM OBJECT TO INT64

data['Size'] = data['Size'].astype(np.int64,errors='ignore')
```

```python
# THE DATA SHOWS INT64 BUT HAS STRING BECAUSE TILL NOW IT CONTAINS NaN

data[data['Size']=="10"].count() 
```

```python
data.isnull().sum()
```

```python
data[['Gender_Category','Product_Line',"Size"]]
```

```python
# HERE IT IS CHECKED THAT WHETHER WE HAVE THE SUFFICIENT VALUES TO COMPUTE THE MEAN/MEDIAN TO SUBSTITE AND FILL THE NULL VALUES.

data.groupby(['Gender_Category','Product_Line'],as_index=True).count()
```

```python
# HERE CATEGORY AND GROUP-WISE UNIQUE VALUES ARE CHECKED TO IDENTIFY THE PATTERN AND DIVERSITY OR ROBUSTNESS OF OUTLIERS.

data.groupby(['Gender_Category','Product_Line'])['Size'].unique()
```

```python
# THE SIZE COLUMN IS FORCED TO NUMERIC FORMAT AND THEN THE NULL VALUES ARE FILLED WITH THE MEDIAN OF GROUP-WISE AVAILABLE VALUES TO SUPPORT THE SUBSTITUTION.

data['Size'] = pd.to_numeric(data['Size'],errors='coerce')
data['Size'] = data.groupby(['Gender_Category','Product_Line'])['Size'].transform(lambda x: x.fillna(x.mean()))
```

```python
# CONVERTING THE DATATYPE OF THE SIZE COLUMN FROM OBJECT TO INT64

data['Size'] = data['Size'].astype(np.int64)
```

```python
data['Size'].isnull().sum()
```

```python
data.info()
```

### CLEANING AND PREPROCESSING THE UNITS_SOLD COLUMN

```python
# THE UNIT SOLD COLUMN OTHER THAN THE NULL VALUES ALSO HAD THE 0.0 VALUES

data[data['Units_Sold']==0]
```

```python
data['Units_Sold'].value_counts()
```

```python
# REPLACING THE -VE VALUES WITH +VE ONES

data['Units_Sold'].replace(-1.0,1.0,inplace=True)
```

```python
data
```

```python
# CHECKING THE CATEGORY-WISE (GROUP-WISE) MEDIAN 

data.groupby(['Gender_Category','Product_Name'])['Units_Sold'].median()
```

```python
# CHECKING THE GROUP-WISE UNIQUE VALUES TO SEE OUTLIERS AND DIVERSITY AMONG THE VALUES PER EACH GROUP.

data.groupby(['Gender_Category','Product_Name'])['Units_Sold'].unique()
```

```python
# AFTER FINDING SUFFICIENT VALUES, THE 0.0 VALUES WAS SUBSTITUTED WITH GROUP-WISE MEADIAN

data['Units_Sold'] = data.groupby(['Gender_Category','Product_Name'])['Units_Sold'].transform(lambda x: x.replace(0.0,x[x!=0.0].median()))
```

```python
# CHANGING THE THE DATA TYPE TO NUMERICAL FORMAT

data["Units_Sold"] = pd.to_numeric(data["Units_Sold"],errors='coerce',downcast='integer')
```

```python
data["Units_Sold"].value_counts()
```

```python
# CONVERTING THE MEDIAN FLOAT VALUES TO ABSOLUTE INTEGER VALUES.

data["Units_Sold"].replace(1.5,2,inplace=True)
data["Units_Sold"].replace(2.5,3,inplace=True)
```

```python
data["Units_Sold"].isnull().sum()
```

```python
# REPLACING THE NULL VALUES WITH THE GROUP-WISE MEDIAN BUT AFTER THE 0.0 VALUES ARE ALREADY REPLACED TO INCREASE THE ACCURACY

data['Units_Sold'] = data.groupby(['Gender_Category','Product_Name'])['Units_Sold'].transform(lambda x: x.fillna(x.median()))
```

```python
data['Units_Sold'].isnull().sum()
```

```python
# CHANGING DATA TYPE TO INT64

data['Units_Sold'] = data['Units_Sold'].astype(np.int64)
```

### CLEANING AND PREPROCESSING THE ORDER_DATE COLUMN

```python
# SORTING THE DATAFRAME W.R.T. "ORDER_DATE" BECUASE WE DECIDED TO FILL IT WITH FORWARD FILL METHOD.

data.sort_values(by='Order_Date',ascending=True,inplace=True)
```

```python
# CHECKING NULL VALUES IN THE COLUMN

data['Order_Date'].isnull().sum()
```

```python
# CHECKING THE DIVERSITY IN THE VALUES. (EVENTUALLY TOLD US ABOUT THE MIXED FORMAT OF THE VALUES STORED)

data['Order_Date'].value_counts()
```

```python
# USING THE FORWARD FILL METHOD TO FILL THE NULL VALUES BECAUSE IT IS NOT POSSIBLE TO FILL IT WITH ANY STATISTICAL APPROACH/MEASURE.

data['Order_Date'].fillna(method='ffill',inplace=True)
```

```python
# SOME PREPROCESSING BEFORE MAKING THE VALUES INTO SAME FORMAT SO THAT LIBRARIES ARE GIVEN A SAME SORT OF INPUT AND IT CAN CONVERTED INTO DATETIME FORMAT.

data['Order_Date'].replace('/','-',inplace=True)
```

```python
# CHANGING THE DATATYPE FROM "OBJECT" TO "DATETIME" TO USE THE FEATURES ASSOCIATED WITH IT.

data['Order_Date'] = pd.to_datetime(data['Order_Date'],errors='coerce',format='mixed')
```

```python
# CHANGING THE FORMAT OF THE DATE TO DD/MM/YY

data['Order_Date']=data['Order_Date'].dt.strftime('%d-%m-%Y')
```
